# Churn Rate Analysis And Ploting:


```python
yes_churn = df.groupby(['Churn','tenure']).count().transpose()['Yes']
```

```python
no_churn = df.groupby(['Churn','tenure']).count().transpose()['No']
```

```python
churn_rate = 100*yes_churn/(no_churn+yes_churn)
```

```python
churn_rate.T['customerID']
```

```python
churn_rate.T['customerID'].plot()
```

```python
# Defining Tenure Cohorts
def cohort(tenure):
    if tenure <13:
        return '0-12 Months'
    elif tenure < 25:
        return '12-24 Months'
    elif tenure < 49 :
        return '24-48 Months'
    else:
        return 'Over 48 Months'
```

```python
# Applying Cohort Function

df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
# Visualizing Monthly vs. Total Charges with Tenure Cohort

sns.scatterplot(data=df,x='MonthlyCharges',y='TotalCharges',hue='Tenure Cohort',alpha=0.4)
```

```python
sns.countplot(data=df,x='Tenure Cohort',hue='Churn')
```

```python
sns.catplot(data=df,x='Tenure Cohort',hue="Churn",kind='count',col='Contract')
```

# Data Preprocessing:

```python
X = df.drop(["Churn",'customerID'],axis=1)
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
y = df["Churn"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

# Decision Tree Model:


```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=6)
```

```python
dt.fit(X_train,y_train)
```

# Predictions and Evaluation:

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
preds = dt.predict(X_test)
```

```python
print(classification_report(y_test,preds))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,preds)
```

```python
imp_feat = pd.DataFrame(data=dt.feature_importances_,index=X_train.columns,columns=['feature Importance'])
```

```python
imp_feat = imp_feat.sort_values(by='feature Importance')
```

```python
imp_feat = imp_feat[imp_feat['feature Importance']>0]
```

```python
imp_feat
```

```python
# Visualizing Feature Importance

plt.figure(figsize=(12,8))
sns.barplot(data=imp_feat,x=imp_feat.index,y='feature Importance')
plt.xticks(rotation=90);
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(dt);
```
